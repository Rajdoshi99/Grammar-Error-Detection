```python
import nltk
bigram=pd.Series(nltk.ngrams([i.split() for i in result.text[153181:]], 2))

```

```python
bigram=[]
for sentence in result.text[153181:253181]:
  bigram.extend(list(nltk.ngrams(sentence.split(),2)))
```

```python
trigram=[]
for i in result.text[153181:253181]:
  trigram.extend(list(nltk.ngrams(i.split(),3)))

#freq_tri = FreqDist(trigram)
```

```python
#fourgram
fourgram=[]
for i in result.text[153181:253181]:
  fourgram.extend(list(nltk.ngrams(i.split(),4)))

```

```python
from nltk import FreqDist
freq_bi = FreqDist(bigram)
freq_tri = FreqDist(trigram)
freq_four = FreqDist(fourgram)
```

```python
from collections import defaultdict
from collections import Counter
import random
d = freq_bi
def noob(x,y):
  count=0
  #print(x)
  #print(y)
  flag=0
  for a, b, c in freq_tri:
    if a==x and b==y:
      #print([a,b,c],end="--->")
      #print(freq_tri[a,b,c])
      flag=1
      count+=1
      if count==10:
        break
      return [a,b,c]
  return -1
  
    
```

```python
from collections import defaultdict
from collections import Counter
import random
d = freq_tri
def noob1(x,y,z):
  count=0
  #print(x)
  #print(y)
  flag=0
  for a, b, c,d in freq_four:
    if a==x and b==y and c==z:
      print([a,b,c,d],end="--->")
      #print(freq_four[a,b,c,d])
      flag=1
      count+=1
      if count==10:
        break
      return [a,b,c,d]
  return -1
  
```

```python
answer=[]
for z in result.text[:1000]:
  a=z.split()
  #print(a)
 
  flagz=0
  for x in range(len(a)-1):
    #print(noob(a[x],a[x+1]))
    if noob(a[x],a[x+1])==-1:
      print("GRAMMAR ERORR")
      flagz=1
      break
  if flagz==0:
    for x in range(len(a)-2):
      if noob1(a[x],a[x+1],a[x+2])==-1:
        flagz=1
        print("GRAMMAR ERORR")
        break
  if flagz==0:
    print("THE SENTENCE IS CORRECT")
    answer.append(0)
  else:
    answer.append(1)
```

```python
for x in range(len(answer)):
  if answer[x]==1:
    answer[x]=0
  elif answer[x]==0:
    answer[x]=1
```
